the retrieved insights using “Matplotlib” modules. Visualize hidden insights using appropriate
plots (graphs) [Usage of line plot and scatter plot are mandatory]

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load your dataset from the CSV file (replace 'your_dataset.csv' with the actual file path)
df = pd.read_csv('banking_management_dataset.csv')

# Scatter Plot
plt.figure(figsize=(12, 6))
plt.scatter(df['MonthlyTransactions'], df['LoanAmount'], alpha=0.5)
plt.xlabel('Monthly Transactions')
plt.ylabel('Loan Amount')
plt.title('Monthly Transactions vs. Loan Amount')
plt.show()

```

```python
# Line Plot
sns.lineplot(df[['Age','LoanAmount']].groupby('Age').sum()).plot()
```

```python
# Bar Graph
plt.figure(figsize=(12, 6))
sns.barplot(x='AccountType', y='Balance', data=df, errorbar=None)
plt.xlabel('Account Type')
plt.ylabel('Average Balance')
plt.title('Average Balance by Account Type')
plt.show()
```

```python
# Line Graph
plt.figure(figsize=(12, 6))
plt.plot(df['Age'], df['Balance'], 'bo', alpha=0.8)
plt.xlabel('Age')
plt.ylabel('Balance')
plt.title('Age vs. Balance')
plt.show()
```
